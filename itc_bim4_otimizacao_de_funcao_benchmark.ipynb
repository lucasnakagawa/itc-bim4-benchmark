**1)	Implemente a função benchmark escolhida.**

**A Egg Crate Function**

```
# Egg Crate Function
import numpy as np

def eggCrateFunction(x, y):
  sinX = np.sin(x)
  sinY = np.sin(y)
  return x**2 + y**2 + (25 * (sinX**2 + sinY**2))
```

**2)	Implemente o algoritmo de busca aleatória.**

```
import random


# Quantidade de tentativas de gerar uma nova solução
iterations = 100

# listas com os dados para gerar o gráfico
solutions = []
attempts = []
valuesX = []
valuesY = []

# contador de tentativas
count = 1

# Gera a solução inicial aleatoriamente respeitando os limites 
x = random.uniform(-5, 5)
y = random.uniform(-5, 5)

# calcula a qualidade da solucao
solution = eggCrateFunction(x, y)

# salva os dados para geração do gráfico
solutions.append(solution)
valuesX.append(x)
valuesY.append(y)
attempts.append(0)


# enquanto houver tentativas
while count < iterations:
    # geracao de uma nova solucao aleatoriamente, respeitando os limites
    tempX = random.uniform(-5, 5)
    tempY = random.uniform(-5, 5)
  
    # Calcula a qualidade da solução gerada
    tempSolution = eggCrateFunction(tempX, tempY)
    
    # se a nova solucao for melhor, atualiza
    if (tempSolution < solution):
        solution = tempSolution
        x = tempX
        y = tempY
    
    # salva os dados para geração do gráfico
    solutions.append(solution)
    attempts.append(count)
    valuesX.append(x)
    valuesY.append(y)
    
    # contabiliza a tentativa
    count = count + 1

#############################################
## 4)	Realize a otimização da função benchmark com os dois algoritmos e gere o gráfico do processo.
#############################################
# Importação da biblioteca MatplotLib

y = solutions
x = attempts

print('x = ', x)
print('y = ', y)

print('Valores x = ', valuesX)
print('Valores y = ', valuesY)

# Biblioteca matplotlib
import matplotlib
matplotlib.use('Agg')
# Gera os gráficos
import matplotlib.pyplot as plt

# Utilizado para cálculos
import numpy as np

# Utilizado para exibir as imagens
from IPython.display import Image

# Dados que serão utilizados


# Gera o gráfico de linhas
plt.plot(x, y )

# Coloca o título
plt.title("Desempenho da busca aleatoria")

plt.ylabel("Valor da funcao objetivo")
plt.xlabel("Iteracoes")

# Salva a figura em um arquivo
plt.savefig('grafico_de_linha')

plt.close('all')

# Exibe a imagem
Image('grafico_de_linha.png')
```

**3)	Implemente o algoritmo por método gradiente. Realize modificações no algoritmo.**

```
#############################################
## Método de Gradiente 
#############################################

import random

# Quantidade de tentativas de gerar uma nova solução
iterations = 10000

# listas com os dados para gerar o gráfico
solutions = []
attempts = []
valuesX = []
valuesY = []

# contador de tentativas
count = 1

# Gera a solução inicial aleatoriamente respeitando os limites 
x = random.uniform(-5, 5)
y = random.uniform(-5, 5)

# calcula a qualidade da solucao
solution = eggCrateFunction(x, x)

# salva os dados para geração do gráfico
solutions.append(solution)
valuesX.append(x)
valuesY.append(y)
attempts.append(0)

# enquanto houver tentativas
while count < iterations:
    # geracao de uma nova solucao
    ###################################
    # aplicando uma pequena modificação
    ###################################
    tempX = x + random.uniform(-0.05, 0.05)
    tempY = y + random.uniform(-0.05, 0.05)

    if (tempX > 5 or tempX < -5):
        tempX = random.uniform(-5, 5)

    if (tempY > 5 or tempY < -5):
        tempY = random.uniform(-5, 5)
    
    # Calcula a qualidade da solução gerada
    tempSolution = eggCrateFunction(tempX, tempY)
    
    # se a nova solucao for melhor, atualiza
    if (tempSolution < solution):
        solution = tempSolution
        x = tempX
        y = tempY
    
    # salva os dados para geração do gráfico
    solutions.append(solution)
    attempts.append(count)
    valuesX.append(x)
    valuesY.append(y)
    
    # contabiliza a tentativa
    count = count + 1

#############################################
## 4)	Realize a otimização da função benchmark com os dois algoritmos e gere o gráfico do processo.
#############################################
# Importação da biblioteca MatplotLib

y = solutions
x = attempts

print('x = ', x)
print('y = ', y)

print('Valores x = ', valuesX)
print('Valores y = ', valuesY)

# Biblioteca matplotlib
import matplotlib
matplotlib.use('Agg')
# Gera os gráficos
import matplotlib.pyplot as plt

# Utilizado para cálculos
import numpy as np

# Utilizado para exibir as imagens
from IPython.display import Image

# Dados que serão utilizados


# Gera o gráfico de linhas
plt.plot(x, y )

# Coloca o título
plt.title("Desempenho da busca por método de gradiente")

plt.ylabel("Valor da funcao objetivo")
plt.xlabel("Iteracoes")

# Salva a figura em um arquivo
plt.savefig('grafico_de_linha')

plt.close('all')

# Exibe a imagem
Image('grafico_de_linha.png')
```
